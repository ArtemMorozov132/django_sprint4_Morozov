```python
import numpy as np

list1_multiplied = [676, 144, 16]
list2_modified = [13, 6, 2, 18, 5]
print("ПОДЗАДАЧА 4. Массивы")
array1 = np.array(list1_multiplied)
array2 = np.array(list2_modified)

print(f"Массив, формируемый из значений изменённого списка list1:  {array1}")
print(f"Массив, формируемый из значений изменённого списка list2:  {array2}")

l1_ext = list1_multiplied.copy()
l2_ext = list2_modified.copy()
max_len = max(len(l1_ext), len(l2_ext))

l1_ext.extend([0] * (max_len - len(l1_ext)))

sum_result = list(map(lambda x, y: x + y, l1_ext, l2_ext))
print(f"Поэлементная сумма двух массивов: {sum_result}")

combined = list(zip(l1_ext, l2_ext))
print(f"Список, совмещённый из двух : {combined}")

array2_final = array2 + 5
print(f"Значения второго массива увеличены на 5 : {array2_final}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random

variant = 8
w = 1.85
step = 0.231
fio = "Морозов Артем"
group = "группа 15.27д-ПИ02/24б"

x = np.arange(-w * variant, w * variant, step)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh_func(x):
    return 2 / (1 + np.exp(-2 * x)) - 1

def relu(x):
    return np.maximum(0, x)

color_sigmoid = "magenta"
color_tanh = (random.random(), random.random(), random.random())
color_relu = "orange"


#1й вариант вывода
plt.figure(figsize=(8, 10))
plt.subplot(2, 1, 1)
plt.plot(x, sigmoid(x), label='sigmoid', color=color_sigmoid, linestyle='-') # Сплошная
plt.plot(x, tanh_func(x), label='tanh', color=color_tanh, linestyle='--')     # Пунктир
plt.title(f"Вариант № {variant}")
plt.xlabel(fio)
plt.ylabel(group)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)

plt.subplot(2, 1, 2)
plt.plot(x, relu(x), label='ReLU', color=color_relu, linestyle='-.')
plt.xlabel(fio)
plt.ylabel(group)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()


#2й вариант вывода
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), label='sigmoid', color=color_sigmoid, linestyle='-')
plt.plot(x, tanh_func(x), label='tanh', color=color_tanh, linestyle='--')
plt.title(f"Вариант № {variant} (Вариант вывода 2)")
plt.xlabel(fio)
plt.ylabel(group)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)

plt.subplot(2, 2, 4)
plt.plot(x, relu(x), label='ReLU', color=color_relu, linestyle='-.')
plt.xlabel(fio)
plt.ylabel(group)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

fio = "Морозов Артем"
group = "группа 15.27д-ПИ02/24б"
variant = 8
X = np.array([1, -1, 3, 2])
Y_true = np.array([1, 1.5])
# Веса
W = np.array([[-0.2, -0.2, 0.2, 0.2],
              [0.4, 0.4, -0.4, -0.4]])
lr = 0.01

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def calculate_net(X, W):
    z = np.dot(W, X)
    V = sigmoid(z)
    return z, V

# --- ПОДЗАДАЧА 1 ---
print("Задача 3.3. Нейронные сети")
print("ПОДЗАДАЧА 1. Реализация на Python задачи 2.1")
print(f"Студент: {fio}, {group}, Вариант {variant}")
print(f"Дата выполнения работы: {datetime.now()}\n")

print(f"Входной слой: {X}")
print(f"Выходной слой: {Y_true}")
print(f"Веса: {W.tolist()}\n")

z, V = calculate_net(X, W)
error = np.sum(0.5 * (Y_true - V)**2)

print(f"Суммарное входное воздействие для нейрона V1: {z[0]:.1f}  Значение нейрона V1: {V[0]:.6f}")
print(f"Суммарное входное воздействие для нейрона V2: {z[1]:.1f}  Значение нейрона V2: {V[1]:.6f}")
print(f"\nОшибка сети: {error:.6f}")
print("-" * 50)

# --- ПОДЗАДАЧА 2 (Backpropagation) ---
print("\nЗадача 4.3. Нейронные сети")
print("ПОДЗАДАЧА 2. Реализация на Python задачи 3.3 Backpropagation")
print(f"Студент: {fio}, {group}, Вариант {variant}")
print(f"Дата выполнения работы: {datetime.now()}\n")

print(f"Первоначально рассчитанное значение нейрона V1: {V[0]:.6f}")
print(f"Первоначально рассчитанное значение нейрона V2: {V[1]:.6f}")
print(f"Первоначально рассчитанная ошибка: {error:.6f}")
print(f"Ошибка в виде массива [{error:.7f}]\n")

errors_history = []
for i in range(100):
    z, V = calculate_net(X, W)
    current_error = np.sum(0.5 * (Y_true - V)**2)
    errors_history.append(current_error)
    d_error_V = -(Y_true - V)
    d_V_z = V * (1 - V)
    delta = d_error_V * d_V_z
    W = W - lr * np.outer(delta, X)

    if i < 4 or i == 99:
        print(f"Новые веса на ( {i} ) итерации: {W[0].tolist()} {W[1].tolist()}")
        print(f"Ошибка сети ( {i} ): {current_error:.14f}\n")

plt.figure(figsize=(10, 5))
plt.plot(range(100), errors_history, color='red')
plt.title(f"{fio}, {group}, Вариант № {variant}")
plt.xlabel("Цикл обучения")
plt.ylabel("Ошибка сети")
plt.grid(True, alpha=0.3)
plt.show()

print(f"НОВАЯ ошибка сети после 100 циклов: {errors_history[-1]:.8f}")
```
